```python
def fun_max():
    names = ("Vinit", "Rucha", "Bal", "Rot")

    current_max = 0
    list_longest_name = ""
    
    for name in names:

        if len(name) > current_max:
            list_longest_name = name
            current_max = len(name)
        elif len(name) == current_max:
            list_longest_name += ", " + name
    
    print("Good morning", list_longest_name, "!")
```

```python
fun_max()
```

```python
def fun7(x,y):
    a = []

    #print(min(len(x), len(y)))
    
    loop = min(len(x), len(y))

    for i in range(loop):
        #print(i)

        a.append(x[i]+y[i])
    print(a)
```
